```python
import os
```

```python
output_folder = "saved_articles"
os.makedirs(output_folder, exist_ok=True)
```

```python
# add more urls here
urls = [
    "https://www.elle.com/fashion/a29845910/jennifer-lopez-coach-global-face/",
    "https://wwd.com/feature/jennifer-lopez-named-global-ambassador-of-coach-1203373789/",
    "https://www.lifestyleasia.com/sg/style/fashion/coach-holiday-gift-guide-2019/",
    "http://www.preview.ph/fashion/best-sellers-coach-bags-philippines-2019-a1879-20191115",
    "https://www.hotelnewsresource.com/article108161.html"
]
```

```python
for i, url in enumerate(urls, 1):
  try:
    article = Article(url)
    article.download()
    article.parse()

    title = article.title
    text = article.text

    safe_title = "".join(c if c.isalnum() else "_" for c in title)
    filename = os.path.join(output_folder, f"{safe_title}.txt")

    with open(filename, "w", encoding="utf-8") as file:
      file.write(f"{article.title}\n\n{text}")

    print(f"[✓] Saved: {filename}")

  except Exception as e:
    print(f"[X] Failed to save article from {url}: {e}")
```
